**Dataset Description**
The Iris dataset consists of 150 samples from three different species of the Iris flower:

- Iris-setosa

- Iris-versicolor

- Iris-virginica

Each sample has 4 features (numerical measurements in centimeters):

- SepalLengthCm – Length of the sepal

- SepalWidthCm – Width of the sepal

- PetalLengthCm – Length of the petal

- PetalWidthCm – Width of the petal

Additionally, there is a categorical column:

Species – The type of Iris flower (target class)

This dataset is well-suited for practicing:

- Data loading and cleaning 

- Summary statistics

- Data visualization and pattern recognition



```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set plot styles
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Load the Iris Dataset
iris = pd.read_csv('/kaggle/input/iris-dataset/IRIS DATASET.csv')
iris.head()
```

```python
#  Explore Dataset Structure
print("Dataset Shape:", iris.shape)       # Rows x Columns
print("Column Names:", iris.columns)      # Feature Names
print("\nFirst 5 Records:\n", iris.head())  # Peek at data
```

```python
# Summary Statistics
print("\nStatistical Summary:\n", iris.describe())
```

```python
#  Scatter Plot: Sepal Length vs. Petal Length
sns.scatterplot(data=iris, x="SepalLengthCm", y="PetalLengthCm", hue="Species")
plt.title("Sepal Length vs. Petal Length by Species")
plt.show()
```

```python
# Histogram: Petal Width Distribution
sns.histplot(iris["PetalWidthCm"], kde=True, bins=20, color="blue")
plt.title("Distribution of Petal Width")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box Plot: Sepal Width by Species
sns.boxplot(x="Species", y="SepalWidthCm", data=iris, palette="pastel")
plt.title("Sepal Width by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.show()
```
